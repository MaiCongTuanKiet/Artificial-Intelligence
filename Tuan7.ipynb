```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

url = 'mtcars.csv'
data = pd.read_csv(url)

X = data.drop(data.columns[[0, 2]], axis=1)
y = data['mpg']


my_range = np.arange(1.5, 5.5, 0.1)
dem_range = np.arange(2, 11)

for degree in dem_range:
    
    a_range = np.arange(1, degree + 1)

  
    X_poly = pd.DataFrame([X['wt'] ** i for i in a_range]).T

  
    model = linear_model.LinearRegression()
    model.fit(X_poly, y)

   
    predictor = pd.DataFrame([my_range ** i for i in a_range]).T
    y_pred = model.predict(predictor)
 
 
    plt.figure()
    plt.scatter(X['wt'], y, color='black')
    plt.plot(my_range, y_pred, color='blue')
    plt.title('Polynomial degree: ' + str(degree))
    plt.xlabel('wt')
    plt.ylabel('mpg')
    plt.xlim(0, 9)

plt.show()
```

```python

from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
corr = data.corr()
sns.heatmap(corr,annot = True)
```

```python
X = data.drop(['State','Profit'],axis=True)
y = data['Profit']
X.head()
     
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape) 
```

```python
regression = linear_model.LinearRegression()
regression.fit(X_train, y_train)
y_pred = regression.predict(X_test.round(1))
calculation = pd.DataFrame(np.c_[y_test,y_pred],columns=['Original Profit','Predicted Profit'])
print(regression.score(X_train,y_train))
print(regression.score(X_test,y_test))
```
